```python
# 데이터프레임 새로 만들기
# 데이터프레임 이름 : df_month_age
df_month_age = df_20TABLE_org[["수진자연령","요양개시일자","심결본인부담금"]]
```

```python
# 9세이하, 10대, 20대, ~~ 90대, 100세 이상

# 9세 이하
df_month_age[(df_month_age["수진자연령"] <= 9)]

# 10대
df_month_age[(df_month_age["수진자연령"] >= 10 & (df_month_age[(df_month_age["수진자연령"] < 20)]

# 20대
df_month_age[(df_month_age["수진자연령"] >= 20 & (df_month_age[(df_month_age["수진자연령"] < 30)]

# 30대
df_month_age[(df_month_age["수진자연령"] >= 30 & (df_month_age[(df_month_age["수진자연령"] < 40)]

# 40대
df_month_age[(df_month_age["수진자연령"] >= 40 & (df_month_age[(df_month_age["수진자연령"] < 50)]

# 50대
df_month_age[(df_month_age["수진자연령"] >= 50 & (df_month_age[(df_month_age["수진자연령"] < 60)]

# 60대
df_month_age[(df_month_age["수진자연령"] >= 60 & (df_month_age[(df_month_age["수진자연령"] < 70)]

# 70대
df_month_age[(df_month_age["수진자연령"] >= 70 & (df_month_age[(df_month_age["수진자연령"] < 80)]

# 80대
df_month_age[(df_month_age["수진자연령"] >= 80 & (df_month_age[(df_month_age["수진자연령"] < 90)]

# 90대
df_month_age[(df_month_age["수진자연령"] >= 90 & (df_month_age[(df_month_age["수진자연령"] < 100)]

# 100세 이상
df_month_age[(df_month_age["수진자연령"] >= 100)]
```

```python
# 9세 이하 만들기
df_under9 = df_month_age[(df_month_age["수진자연령"] <= 9)]

# index 재배열
df_under9 = df_under9.reset_index(drop = True)

# "연령대" 컬럼 추가 후 값은 모두 9세 이하 넣기
df_under9["연령대"] = "9세이하"

df_under9.head(2)
```
