### 30년간 출산율 추이분석

```python
# csv 파일 가져오기
file_path = "C:/Users/admin/StudyAnaconda/team_project/김재홍 데이터/데이터프레임/birth_30years(kosis).csv" # 파일경로
birth_30 = pd.read_csv(file_path)
```

```python
birth_30
```

```python
# 시점 열을 인덱스로 설정하기
birth_30.set_index("시점")
```

```python
# 필요없는 Unnamed: 0 열을 삭제하기
birth30 = birth_30.drop(columns=["Unnamed: 0"])
```

```python
birth30
```

```python
# 시점 기준으로 인덱스 변경
b30 = birth30.set_index("시점")
```

```python
b30
```

```python
# 맷플롯립 라이브러리 가져오기
import matplotlib as mpl
import matplotlib.pylab as plt
```

```python
# 한글깨짐 방지
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.rc('figure', titlesize=50)
```

```python
# 꺽은선 그래프로 그리기
b30.plot()
```

```python
b30.info()
```

```python
# 30년간 출생아수와 합계출산율 분석

plt.subplot(3, 2, 1) # 서브플롯 위치 좌표
plt.plot(b30.iloc[: , [0]], 'r') # 데이터프레임 행 추출, 선 색깔(빨강)
plt.rcParams['figure.figsize'] = (30, 20) # 그래프 크기
plt.rcParams['lines.linewidth'] = 2 # 선 너비
plt.rc('font', family = 'Malgun Gothic', size = 10) # 폰트와 글씨 크기
plt.ylabel('출생아수(명)') # y축 제목
plt.title('30년간 출산아수 추이분석') # 그래프 제목
plt.grid(True) # 그리드 표시


plt.subplot(3, 2, 2) # 서브플롯 위치 좌표
plt.plot(b30.iloc[: , [1]], 'g') # 데이터프레임 행 추출, 선 색깔(녹색)
plt.rcParams['figure.figsize'] = (30, 20) # 그래프 크기
plt.rcParams['lines.linewidth'] = 2 # 선 너비
plt.rc('font', family = 'Malgun Gothic', size = 10) # 폰트와 글씨 크기
plt.ylabel('자연증가건수(명)') # y축 제목
plt.title('30년간 자연증가건수 추이') # 그래프 제목
plt.grid(True) # 그리드 표시

plt.subplot(3, 2, 3) # 서브플롯 위치 좌표
plt.plot(b30.iloc[: , [2]], 'y') # 데이터프레임 행 추출, 선 색깔(노랑)
plt.rcParams['figure.figsize'] = (30, 20) # 그래프 크기
plt.rcParams['lines.linewidth'] = 2 # 선 너비
plt.rc('font', family = 'Malgun Gothic', size = 10) # 폰트와 글씨 크기
plt.ylabel('조출생률(천명당)') # y축 제목
plt.title('30년간 조출생율 추이') # 그래프 제목
plt.grid(True) # 그리드 표시

plt.subplot(3, 2, 4) # 서브플롯 위치 좌표
plt.plot(b30.iloc[: , [3]], '#7FFFD4') # 데이터프레임 행 추출, 선 색깔(서브마린 좌표)
plt.rcParams['figure.figsize'] = (30, 20) # 그래프 크기
plt.rcParams['lines.linewidth'] = 2 # 선 너비
plt.rc('font', family = 'Malgun Gothic', size = 10) # 폰트와 글씨 크기
plt.ylabel('자연증가율(천명당)') # y축 제목
plt.title('30년간 자연증가율 추이') # 그래프 제목
plt.grid(True) # 그리드 표시

plt.subplot(3, 2, 5) # 서브플롯 위치 좌표
plt.plot(b30.iloc[: , [5]], 'c') # 데이터프레임 행 추출, 선 색깔(cyan)
plt.rcParams['figure.figsize'] = (30, 20) # 그래프 크기
plt.rcParams['lines.linewidth'] = 2 # 선 너비
plt.rc('font', family = 'Malgun Gothic', size = 10) # 폰트와 글씨 크기
plt.ylabel('출생성비(명)') # y축 제목
plt.title('30년간 출생성비 추이') # 그래프 제목
plt.grid(True) # 그리드 표시

plt.subplot(3, 2, 6) # 서브플롯 위치 좌표
plt.plot(b30.iloc[: , [4]], 'b') # 데이터프레임 행 추출, 선 색깔(파랑)
plt.rcParams['figure.figsize'] = (30, 20) # 그래프 크기
plt.rcParams['lines.linewidth'] = 2 # 선 너비
plt.rc('font', family = 'Malgun Gothic', size = 10) # 폰트와 글씨 크기
plt.ylabel('합계출산율(명)') # y축 제목
plt.title('30년간 합계출산율 추이') # 그래프 제목
plt.grid(True) # 그리드 표시


# png파일 저장

plt.savefig('birth_in_30years.png')
plt.savefig('birth_in_30years.png', dpi=50)
plt.savefig('birth_in_30years.png', dpi=200)
```
