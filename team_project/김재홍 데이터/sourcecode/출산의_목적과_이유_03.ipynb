## 저출산의 원인 분석

#### 현재 대한민국의 출산율은 시간이 갈수록 떨어지고 있다. 고령화사회로 급히 변해가는 과정 중에 가장 근본적인 문제 저출산에 대해 분석해보고 이유를 도출해보려고 한다. 

```python
import pandas as pd
```

```python
# 파일 불러오기
# 데이터 출처: kosis 기혼여성(15~49세)의 자녀 필요성 데이터
file_path = "C:/Users/admin/StudyAnaconda/team_project/김재홍 데이터/데이터프레임/기혼여성_1549세_의_자녀_필요성2018(KOSIS).csv"
df = pd.read_csv(file_path, header =2)
```

## 전처리

```python
# 앞부분만 보기
df.head()
```

```python
# 데이터 타입 확인
df.info()
```

```python
# 컬럼명 제거하기
df2 = df.drop(["특성별(1)"], axis=1)
```

```python
df2
```

```python
# 컬럼명 바꾸기
df3 = df2.rename(columns= {"특성별(2)" : "나이대"})
```

```python
df3.head()
```

```python
# 소계 행 삭제
df4 =  df3.drop(0)
```

```python
df4
```

```python
# 인덱스 설정
birth_reason = df4.set_index("나이대")
```

```python
birth_reason.head()
```

```python
# 필요한 컬럼만 뽑아서 데이터 프레임 만들기
subset1 = birth_reason.loc[:,["꼭 있어야 한다 (%)",
                              "있는 것이 없는 것보다 나을 것이다 (%)",
                              "없어도 무관함 (%)","모르겠음 (%)"]]
```

```python
subset1
```

```python
# 필요한 컬럼만 뽑아서 데이터 프레임 만들기
subset2 = birth_reason.loc[:,["가문(대)를 잇기 위해 (%)",
                              "노후생활을 위해서 (%)",
                              "심리적 만족 (%)",
                              "가정의 행복과 조화 (%)",
                              "기타 (%)"]]
```

```python
subset2
```

```python
# 필요한 컬럼만 뽑아서 데이터 프레임 만들기
subset3 = birth_reason.loc[:,["경제적으로 여유롭게 생활 (%)",
                              "직장생활 지속 희망 (%)",
                              "부부만의 생활 추구 (%)",
                              "자녀가 있으면 자유롭지 못할 것 같아 (%)", 
                              "경제적으로 자녀양육 곤란 (%)",
                              "아이 희망하나 불임 등으로 임신 곤란 (%)",
                              "아이가 행복하게 살기 힘든 사회여서 (%)",
                              "기타 (%).1"]]                             
```

```python
# 컬럼명 바꾸기
subset_reset = subset3.rename(columns={"기타 (%).1":"기타 (%)"})
```

```python
subset_reset
```

```python
import matplotlib.pyplot as plt
```

```python
### 출산의 이유, 목적 분석하고 시각화 하기

# 한글 깨짐 방지 코드
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.rc('figure', titlesize=50)


ax1 = subset1.plot(kind='bar', title='자녀의 필요성', figsize=(10, 10), legend=True, fontsize= 15) # 그래프 종류와 제목, 크기, 범례설정
plt.xticks(rotation= 0) # x축 레이블명 회전
ax1.set_xlabel('나이대', fontsize = 15)    # x축 정보 표시
ax1.set_ylabel('비율', fontsize = 15)     # y축 정보 표시
plt.savefig("출산의 이유와 목적분석1.png") # 이미지 파일 저장

ax2 = subset2.plot(kind='bar', title='자녀가 있어야 한다고 응답한 이유', figsize=(10, 10), legend=True, fontsize= 15) # 그래프 종류와 제목, 크기, 범례설정
plt.xticks(rotation= 0) # x축 레이블명 회전
ax2.set_xlabel('나이대', fontsize = 15)          # x축 정보 표시
ax2.set_ylabel('비율', fontsize = 15)     # y축 정보 표시
plt.savefig("출산의 이유와 목적분석2.png") # 이미지 파일 저장

ax3 = subset_reset.plot(kind='bar', title='자녀가 없어도 된다고 응답한 이유', figsize=(15, 10), legend=True, fontsize= 15) # 그래프 종류와 제목, 크기, 범례설정
plt.xticks(rotation= 0) # x축 레이블명 회전
ax3.set_xlabel('나이대', fontsize=15)          # x축 정보 표시
ax3.set_ylabel('비율', fontsize= 15)     # y축 정보 표시
plt.savefig("출산의 이유와 목적분석3.png") # 이미지 파일 저장

# 그래프 제목
ax1.set_title('자녀의 필요성')
ax2.set_title("자녀가 있어야 한다고 응답한 이유")
ax3.set_title("자녀가 없어도 된다고 응답한 이유")
```
