```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./files_new/auto_mpg_v2.csv")
```

```python
df
```

```python
# 지도학습 모델 : 단순회귀분석(선형회귀모델, 선형회귀예측모델)
# - 종속변수가 있음
# - 직선의 방정식을 사용 y = ax + b
# - y : 종속변수(정답)
# - x : 독립변수
# - a : 기울기
# - b : y절편

# - 학습 : fit함수
```

```python
# 사용할 변수들 : mpg, cylinders, horsepower, weight
ndf = df[["mpg","cylinders","horsepower","weight"]]
ndf
```

```python
# 산점도 그래프 그리기 - 종속변수 mpg, 독립변수 : weight
fig = plt.figure(figsize = (10, 5))

# 1번째 그래프 공간 만들기
ax1 = fig.add_subplot(1,2,1)
sns.regplot(x = "weight",# 독립변수
            y = "mpg", # 종속변수
            data = ndf,
            color = "red",     # 그래프 색깔
            ax = ax1,
            fit_reg = True)

# 2번째 그래프 공간 만들기
ax2 = fig.add_subplot(1,2,2)
sns.regplot(x = "weight",# 독립변수
            y = "mpg", # 종속변수
            data = ndf,
            color = "blue",     # 그래프 색깔
            ax = ax2,
            fit_reg = False)

plt.show()
plt.close() # 메모리에서 삭제하기
```

```python
# 해석 
# 중량이 커질수록 연비는 낮아지는 음의 상관관계를 갖는다.
# 연비는 중량에 영향을 미친다
```

```python
# join plot 그리기
joint = sns.jointplot(x = "weight",
                      y = "mpg",
                      kind = "reg",
                      data = ndf)
joint.set_axis_labels(xlabel = "weight",
                      ylabel = "mpg")

# 추세선 없이 그리기
joint = sns.jointplot(x = "weight",
                      y = "mpg",
                      color = "green",
                      data = ndf)
joint.set_axis_labels(xlabel = "weight",
                      ylabel = "mpg")
```

```python
# 산점도 그래프 그리기 - 종속변수 mpg, 독립변수 : weight
fig = plt.figure(figsize = (10, 5))

# 1번째 그래프 공간 만들기
ax1 = fig.add_subplot(1,2,1)
sns.regplot(x = "mpg",# 독립변수
            y = "horsepower", # 종속변수
            data = ndf,
            color = "red",     # 그래프 색깔
            ax = ax1,
            fit_reg = True)

# 2번째 그래프 공간 만들기
ax2 = fig.add_subplot(1,2,2)
sns.regplot(x = "mpg",# 독립변수
            y = "horsepower", # 종속변수
            data = ndf,
            color = "blue",     # 그래프 색깔
            ax = ax2,
            fit_reg = False)

plt.show()
plt.close() # 메모리에서 삭제하기
```

```python
# join plot 그리기
joint = sns.jointplot(x = "mpg",
                      y = "horsepower",
                      kind = "reg",
                      data = ndf)
joint.set_axis_labels(xlabel = "mpg",
                      ylabel = "horsepower")

# 추세선 없이 그리기
joint = sns.jointplot(x = "mpg",
                      y = "horsepower",
                      color = "green",
                      data = ndf)
joint.set_axis_labels(xlabel = "mpg",
                      ylabel = "horsepower")
```
