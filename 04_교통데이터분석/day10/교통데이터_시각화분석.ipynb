# 데이터 시각화 분석

```python
# 라이브러리와 폰트설정
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
import platform

plt.rc('font', family = 'Malgun Gothic')
```

```python
# 1. 기준월 및 기준일자별 버스 이용량 분석비교
# 사용할 분석요소 : 기준월, 기준일, 승객구분
# 사용할 함수 : 카운트(빈도)
df_bus_card_tot.head(1)
```

```python
# 피봇사용
# 월별/연령대별 수진자연령의 빈도표
pivot_bus_tendency = df_bus_card_tot.pivot_table(values = "승객구분",
                                                    index = "기준월",
                                                    columns = "기준일",
                                                    aggfunc = "count")
pivot_bus_tendency
```

```python
# 결측치 처리하기
df_pivot = pivot_bus_tendency.fillna(0)
df_pivot
```

```python
# 히트맵 그리기

# 그래프 크기 결정
plt.figure(figsize = (16,10))

# 피봇 데이터 넣어서 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = "01", cmap = "rocket_r")

# 타이틀 넣기
plt.title("기준월 및 기준일자별 버스 이용량 분석")

# 보여주기
plt.show()
```

```python
# 기준일별 시간대별 버스 이용량 분석
pivot_bus_tendency = df_bus_card_tot.pivot_table(values = "승객구분",
                                                    index = "기준일",
                                                    columns = "기준시간",
                                                    aggfunc = "count")
pivot_bus_tendency
```

```python
# 결측치 0으로 대체
df_pivot = pivot_bus_tendency.fillna(0)
df_pivot
```

```python
# 히트맵 그리기

# 그래프 크기 결정
plt.figure(figsize = (16, 10))

# 피봇 데이터 넣어서 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r")

# 타이틀 넣기
plt.title("기준일별 시간대별 버스 이용량 분석")

# 보여주기
plt.show()
```

```python
# 기준시간 및 분별 버스 이용량 분석
```

```python
df_pivot = df_bus_card_tot.pivot_table(values = "버스내체류시간(분)",
                                       index = "기준일",
                                       columns = "기준시간",
                                       aggfunc = "mean")
df_pivot
```

```python
# 5. 시간대별 승객구분별 버스내 체류시간(분) 분석
```

```python
df_temp = pd.DataFrame()
df_temp["기준시간"] = df_bus_card_tot["기준시간"]
df_temp["승객구분"] = df_bus_card_tot["승객구분"]
df_temp["버스내체류시간"] = df_bus_card_tot["버스내체류시간(분)"]

df_temp
```

```python
df_temp.drop_duplicates(["승객구분"])
```

```python
df_temp
```

```python
# 시간대별 승객구분별 버스내 체류시간(분) 분석
df_temp2 = (df_temp.groupby(["기준시간", "승객구분"], as_index = False).sum()).\
            sort_values(by = ["버스내체류시간"], ascending = False)
```

```python
# 옆으로 뉘이기
df_temp2.head(10).transpose()
```

```python
fig = plt.figure(figsize = (25, 10))
sns.barplot(x = "기준시간", y = "버스내체류시간", hue = "승객구분", data = df_temp2)
plt.title("시간대별 승객구분별 버스내 체류시간(분) 분석")
plt.show()
```

```python
# 시간 및 승객구분별 빈도분석
fig = plt.figure(figsize = (12, 4))
sns.histplot(data = df_temp2,
             x = "기준시간",
             bins = 30,  # bin 갯수
             kde = True, # 밀도선 보여주기
             hue = "승객구분",
             multiple = 'stack', # dodge
             stat = "density",   # 빈도수
             shrink = 0.6)
plt.title("시간 및 승객구분별 빈도분석")
plt.xlabel("기준시간")
plt.ylabel("rate")
plt.show()
```

```python
# 승하차정류장별 버스내 체류시간(분) 상위 30건 분석 비교
```

```python
df_bus_card_tot
```

```python
df_temp = pd.DataFrame()
df_temp["승하차정류장"] = df_bus_card_tot["승차정류장"] + "-->" + \
                                            df_bus_card_tot["하차정류장"]
df_temp["승객구분"] = df_bus_card_tot["승객구분"]
df_temp["버스내체류시간"] = df_bus_card_tot["버스내체류시간(분)"]

df_temp
```

```python
# 승하차정류장(승차정류장 및 하차정류장)별 버스내체류시간 상위 30건 조회
# 승하차정류장 및 승객구분별로 버스내체류시간의 sum을 구하기
df_temp_sel = (df_temp.groupby(["승하차정류장", "승객구분"],
                               as_index = False).sum()).sort_values(by = ["버스내체류시간"],
                                                                   ascending = False).head(30)
df_temp_sel
```

```python
# 그래프 사이즈 정하기
plt.figure(figsize = (12, 4))

# 데이터 x축, y축 넣기
plt.plot(df_temp_sel["승하차정류장"], df_temp_sel["버스내체류시간"])

# 그래프 제목
plt.title("승하차정류장(승차정류장 및 하차정류장)별 버스내체류시간 상위 30건 분석")

# x축 제목
plt.xlabel("승하차정류장")

# y축 제목
plt.ylabel("버스내체류시간")

# x축 표시값을 90도 수직으로 표현
plt.xticks(rotation = 90)

# 격자 보이기
plt.grid(True)

# 그래프 보여줘~~~~
plt.show()
```
