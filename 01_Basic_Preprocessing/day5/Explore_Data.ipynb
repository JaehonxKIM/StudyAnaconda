## 2. 재구조화

```python
# cut(): 동일한 길이(값의 범위)로 나누기
df1 = pd.cut(df["math"],3)
df1
```

```python
df2 = df["math"].groupby(df1)
```

```python
df2.head(1)
```

```python
# df2에 대한 기초 통계 : agg() 함수 사용
df3 = df2.agg(["count","mean","min","max"])
df3
```

```python
df
```

```python
# qcut()

pd.qcut(df["math"],3)
```

```python
df4 = pd.qcut(df["math"],3, labels = False)
```

```python
df4
```

```python
pd.qcut(df["math"],3)
```

```python
import numpy as np
```

```python
df5 = pd.qcut(df["math"], 3, labels = np.arange(3, 0, -1))
```

```python
# 첫번째값 : 시작값
# 두번째값 끝값 -1
# 세번째값 : 증가값 or 감소값

for i in range(0, 3, 1):
    print(i)
    
print("--------")

for j in range(3, 0, -1):
    print(j)
```

```python
# 첫번째값 : 시작값
# 두번째값 끝값 -1
# 세번째값 : 증가값 or 감소값

for i in np.arange(0, 3, 1):
    print(i)

print("--------")    

for j in np.arange(3, 0, -1):
    print(j)
```

```python
df5
```

```python
df6 = df["math"].groupby(df5)
df6.head(1)
```

```python
# 기초통계 통계조회 
# count , mean, std, min, max  * std 는 중앙값
df7 = df6.agg(["count","mean","std","min","max"])
df7
```
