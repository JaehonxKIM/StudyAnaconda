```python
import pandas as pd
```

```python
# openpyxl 모듈 설치하기 
#! pip install openpyxl

# 엑셀 파일 읽어오기
raw_total = pd.read_excel("./files/3_1_crawling_raw.xlsx")
```

```python
raw_total
```

```python
raw_total["tags"]
```

```python
tags_total = []

# 해시태그를 모두 추출하여 하나의 리스트로 만들기
for tags in raw_total["tags"] :
    # print(tags[2 : -2].split("', '")) # 기준으로 자르기
    split_list = tags[2 : -2].split("', '")
    for tag in split_list :
        tags_total.append(tag) # 리스트에 넣어주기

tags_total        
```

```python
from collections import Counter
```

```python
# 해시태그 별로 빈도 확인 
tag_counts = Counter(tags_total)
tag_counts
```

```python
# 가장 많은 count 상위 50개 조회
tag_counts.most_common(50)
```

```python
# 키워드 정제하기 
# 정제할 키워드에 대한 리스트 만들기
STOPWORDS = ["#일상", "#선팔","#제주도","#jeju", "#반영구",
             "#제주자연눈썹", "#서귀포눈썹문신", "#제주눈썹문신",
             "#소통", "#맞팔"]
```

```python
# for 문을 이용해 리스트에 집어넣기
tag_total_selected = []

for tag in tags_total :
    if tag not in STOPWORDS :
        tag_total_selected.append(tag)

tag_total_selected        
```

```python
tag_total_selected = Counter(tag_total_selected)
tag_total_selected
```

```python
# 라이브러리 가져오기
import matplotlib.pyplot as plt
```

```python
# barplot으로 나타내기
plt.figure(figsize = (10,8))
sns.barplot(x = "counts",
            y = "tags",
            data = tag_counts_df)
```
